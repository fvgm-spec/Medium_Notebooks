#### Parts-of-speech and tagging

```python
#Importing required packages
import nltk
from nltk.corpus import genesis
```

Downloading the genesis corpora

```python
genesis_raw = nltk.corpus.genesis.raw()
genesis_text = nltk.word_tokenize(genesis_raw)
#Getting the first 15 tokens in the text
genesis_tokens = genesis_text[:15]
```

Generating tuples of token-tag for the tokens selected

```python
pos_tags = nltk.pos_tag(genesis_tokens)
pos_tags
```

```python
pos_tags = nltk.pos_tag(genesis_tokens,tagset='universal')
pos_tags
```

```python
nltk.download('tagsets')
```

```python
# Get a list of all the tags used in the corpus
tagset = set(tag for word, tag in pos_tags)
print(tagset)
```

#### Default Tagger

```python
from nltk.corpus import brown
tagged_sents = brown.tagged_sents(categories='science_fiction')
sents = brown.sents(categories='science_fiction')
```

```python
tagged_sents
```

```python
sents
```

```python
tags = [tag for (word, tag) in brown.tagged_words(categories='science_fiction')]
nltk.FreqDist(tags).most_common()
```

```python
raw = 'Turned, the babies at once fell silent, then began to crawl towards those clusters of sleek colours, those shapes so gay and brilliant on the white pages.'
tokens = nltk.word_tokenize(raw)
default_tagger = nltk.DefaultTagger('NN')
default_tagger.tag(tokens)
```

```python
default_tagger.accuracy(tagged_sents)
```

#### Regular Expression Tagger

```python
patterns = [(r'.*ed$', 'VBD'), # simple past
            (r'.*\'s$', 'NN$'), # possessive nouns
            (r'.*es$', 'VBZ'), # 3rd singular present
            (r'.*ing$', 'VBG'), # gerunds
            (r'.*ould$', 'MD'), # modals
            (r'.*s$', 'NNS'), # plural nouns
            (r'^-?[0-9]+(.[0-9]+)?$', 'CD'), # cardinal numbers
            (r'.*', 'NN') # nouns
            ] 
```

```python
regexp_tagger = nltk.RegexpTagger(patterns)
regexp_tagger.tag(sents[7])
```

```python
regexp_tagger.accuracy(tagged_sents)
```

#### Lookup Tagger

```python
fd = nltk.FreqDist(brown.words(categories='science_fiction'))
cfd = nltk.ConditionalFreqDist(brown.tagged_words(categories='science_fiction'))
most_freq_words = fd.keys()
```

```python
most_freq_words
```

```python
likely_tags = dict((word, cfd[word].max()) for word in most_freq_words)
likely_tags
```

```python
baseline_tagger = nltk.UnigramTagger(model=likely_tags)
baseline_tagger.accuracy(tagged_sents)
```
